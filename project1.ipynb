```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
flights.shape
```

matplotlib

```python
yearlist=flights["year"].tolist()
passengerscount=flights["passengers"].tolist()

plt.plot(yearlist,passengerscount,label="number of passengers",marker="*",color="k",markerfacecolor="r",ls="--",lw="3")

plt.ylabel("no.of passengers")
plt.xlabel("year")

plt.xticks(yearlist)
plt.title("passengers per year")
plt.yticks(100,200,300,400)

plt.legend()
plt.show()
```

```python
yearlist=flights["year"].tolist()
passengerscount=flights["passengers"].tolist()

plt.scatter(yearlist,passengerscount,label="number of passengers",marker="*",color="green")

plt.ylabel("no.of passengers")
plt.xlabel("year")

plt.xticks(yearlist)
plt.title("passengers per year")

plt.legend(loc="upper left")
plt.grid(True,ls="--",lw="3")
plt.show()
```

```python
yearlist=flights["year"].tolist()
passengersrange=[100,200,300,400,500,600]

plt.hist(yearlist,passengersrange,label="number of passengers")

plt.xlabel("no.of passengers")
plt.ylabel("year")

plt.xticks(passengersrange)
plt.title("passengers per year")
plt.yticks(yearlist)

plt.legend(loc="upper left")
plt.show()
```

```python
yearlist=flights["year"].tolist()
passengerscount=flights["passengers"].tolist()

plt.pie(yearlist,passengerscount,radius=1.5,startangle=90)
plt.show()
```

```python
from matplotlib import rcParams

rcParams["figure.figsize"]=12,19
sns.lmplot(x="passengers",y="year",data=flights,markers="*",hue="month")
plt.show()
```
